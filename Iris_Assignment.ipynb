```python
import pandas as pd # Tablo verisi ile çalışmak için.
import numpy as np # Matematik ve dizi işlemleri için.
import matplotlib.pyplot as plt # Grafik çizmek için.
%matplotlib inline
# Jupyter Notebook’ta grafiklerin sayfanın içinde görünmesini sağlar.
```

## Task 1

```python
from sklearn.datasets import load_iris
iris = load_iris() # Scikit-Learn içindeki hazır Iris dataseti yükler.
df = pd.DataFrame(iris.data, columns=iris.feature_names) # Yüklenen veriyi pandas tablosuna dönüştürür.
```

```python
df["species"] = iris.target
df.head()
# Çiçek türlerini tabloya ekler ve ilk 5 satırı görüntüler.
```

## Task 2

```python
df["species"] = df["species"].map({0: "setosa", 1: "versicolor", 2: "virginica"})
df.head()
# species sütunundaki sayıları isimlere dönüştürür.
# map fonksiyonu uygulama fonksiyonudur.
```

## Task 3

```python
grup = df.groupby("species") # Tablodaki verileri çiçek türüne göre gruplar.
print("Türlere göre ortalama değerler:")
print(grup.mean()) # Her türün ortalama ölçüleri
```

## Bonus

```python
# define function that accepts a row of data and returns a predicted species
```

```python
# Kural tabanlı tahmin fonksiyonu
def tahmin_et(satir):
    pl = satir["petal length (cm)"]
    pw = satir["petal width (cm)"]

    if pl < 2:
        return "setosa"
    elif pw > 1.8:
        return "virginica"
    else:
        return "versicolor"
```

```python
df["tahmin"] = df.apply(tahmin_et, axis=1)
df[["species", "tahmin"]].head()
# Bu kod sayesinde her çiçek için modelinin yaptığı tahmini görülür ve gerçek tür ile karşılaştırabilir.
```

```python
dogruluk = (df["species"] == df["tahmin"]).mean() # Her satırda gerçek tür ile tahmin edilen tür aynı mı?
print("Kural tabanlı model doğruluğu:", dogruluk)
```
